```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/kedarmhatre/cnx nifty small cap 100 august/29nov24.csv")
df
```

```python
# Group the data by 'Sector' and calculate the average of 'ALPHA NEW' for each sector
sector_avg_alpha = df.groupby('Sector')['ALPHA NEW'].mean()

# Sort the sector averages in descending order
sector_avg_alpha = sector_avg_alpha.sort_values(ascending=False)

# Create a bar plot to visualize the average of 'ALPHA NEW' for each sector
sector_avg_alpha.plot(kind='bar')
plt.xlabel('Sector')
plt.ylabel('Average of ALPHA NEW')
plt.title('Average of ALPHA NEW for Each Sector')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
# Step 1: Calculate the average ALPHA NEW for each sector
sector_avg = df.groupby("Sector")["ALPHA NEW"].mean()

# Step 2: Arrange the sectors in descending order based on the average ALPHA NEW
sorted_sectors = sector_avg.sort_values(ascending=False)

# Step 3: Identify the top 4 sectors
top_4_sectors = sorted_sectors.head(15)

# Step 4: List stocks with ALPHA NEW values above the sector's average
result = {}
for sector in top_4_sectors.index:
    avg_alpha = top_4_sectors[sector]
    stocks_above_avg = df[(df["Sector"] == sector) & (df["ALPHA NEW"] > avg_alpha)]
    result[sector] = stocks_above_avg

# Print the result
for sector, stocks_df in result.items():
    print(f"Sector: {sector} (Average ALPHA NEW: {top_4_sectors[sector]:.2f})")
    print(stocks_df)
    print("\n")
```

```python
result = {}

for sector in top_4_sectors.index:
    avg_alpha = top_4_sectors[sector]
    
    # Select stocks above average ALPHA NEW in the sector
    stocks_above_avg = df[(df["Sector"] == sector) & (df["ALPHA NEW"] > avg_alpha)]
    
    # Sort stocks_above_avg based on ALPHA NEW values in descending order
    sorted_stocks = stocks_above_avg.sort_values(by="ALPHA NEW", ascending=False)
    
    # Select the top 2 companies
    top_2_companies = sorted_stocks.head(2)["NSE CODE"].to_list()
    
    # Store the result
    result[sector] = top_2_companies

# Print the result
for sector, nse_codes in result.items():
    for code in nse_codes:
        print(code)
```
